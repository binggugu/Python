```python
import pandas as pd

date1 = ['A01', 'A02', 'A03', 'A04', 'A05']
date2 = ['201501', '201501', '201501', '201501', '201501']
date3 = ['100', '120', '140', '160', '180']

date_df1 = pd.DataFrame(date1, columns=['col1'])
date_df2 = pd.DataFrame(date2, columns=['col2'])
date_df3 = pd.DataFrame(date3, columns=['col3'])

final = pd.concat([date_df1,date_df2, date_df3], axis = 1) # 합쳐서, 가로로 보여주라
final
```

### 3.1 인덱스 활용하여 조회하기

```python
# 데이터프레임명.iloc[ : , : ]  # 앞에는 행 인덱스 뒤에는 열 인덱스   ‘:’ 만 활용 시 전체
```

```python
# PANDAS 패키지 불러오기
import pandas as pd
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_csv("C:/spark/bin/data/customerdata.csv")

# 컬럼해더 재정의
customerData.columns = ['CUSTID','AVGPRICE','EMI','DEVICECOUNT','PRODUCTAGE','CUSTTYPE']
```

```python
# 인덱스로 조회하기
dfc1 = customerData.iloc[0:1,:]
dfc1
```
