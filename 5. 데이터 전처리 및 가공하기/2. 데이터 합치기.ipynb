### regionid랑 product group by하고 평균을 구해서 컬럼 하나 더 생성
### 원래 테이블이랑 left조인하고 RATIO컬럼 하나 더 생성

```python
import pandas as pd 
import pymysql 
from sqlalchemy import create_engine 

engine = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl') 

regionMasterData = pd.read_sql_query('SELECT * FROM KOPO_REGION_MST', engine) 
selectExample = pd.read_sql_query('SELECT * FROM KOPO_CHANNEL_SEASONALITY_NEW', engine) 

regionMasterData.columns = [x.upper() for x in regionMasterData.columns]
selectExample.columns = [x.upper() for x in selectExample.columns]

print(regionMasterData.head(2))
print(selectExample.head(2))
```

```python
print(len(regionMasterData))
print(len(selectExample))
```

```python
groupData = selectExample.groupby(['REGIONID','PRODUCT'], as_index=False).mean()
print(groupData.head())
print(len(groupData))
```

```python
mergeData = pd.merge(selectExample, groupData,\
        on=["REGIONID","PRODUCT"],suffixes=('_A', '_B'), how="left")
mergeData.head(2)
```

```python
mergeData["RATIO"] = mergeData["QTY_A"]/mergeData["QTY_B"]
mergeData.head(10)
```
