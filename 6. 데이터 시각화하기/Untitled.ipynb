### 2. 1차원 차트 꾸미기

```python
# 차트의 색 및 내용 채우기
```

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

# 커맨드뷰에서 차트 시연
%matplotlib inline
size = 50
#사이즈만큰 랜덤한 데이터 생성 (정규분포: 평균0, 표준편차1)
y=np.random.standard_normal(size)
x=range(size)
# 차트 꾸미기
plt.plot(x,y, 'b', lw=1.5) #차트 선색 및 크기설정
plt.plot(x,y, 'o', ) #차트 내 'o' 표기
plt.grid(True) #그리드 표기
plt.xlabel('index') # x축 라벨
plt.ylabel('value') # y축 라벨
plt.title('Simple plot') # 차트 제목
```
