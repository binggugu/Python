### 1. 차트 종류 변경

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 150개의 2열짜리 정규분포 생성
y=np.random.standard_normal((150,2))

plt.figure(figsize=(7,5))
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plt')
plt.scatter(y[:,0],y[:,1],marker='o', edgecolors='r')
#plt.scatter?
```

### 2. 차원추가 (Scatter 차트)

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 정규분포 50개행 컬럼2개
y=np.random.standard_normal((50,2))
# 0~10까지 랜덤숫자 50개
c= np.random.randint(0,10,len(y))

plt.figure(figsize=(7,5))
plt.scatter(y[:,0],y[:,1],marker='o',c=c)
plt.colorbar()

plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plt')
#plt.scatter?
```
