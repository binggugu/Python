### 1. 객체 차트

```python
import numpy as np
import matplotlib.pyplot as plt
#객체 차트
# 빈캔버스에 추가함
fig = plt.figure()
# 위치 및 너비/높이

# 1~5까지 11개의 점 삽입
x = np.linspace(1,5,11)
y = x**2

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

ax1.plot(x , y)
ax2.plot(y , x)
ax1.set_title('plot1')
ax2.set_title('plot2')
fig
```

```python
import numpy as np
import matplotlib.pyplot as plt
#객체 차트
# 빈캔버스에 추가함
fig = plt.figure()
# 위치 및 너비/높이

# 1~5까지 11개의 점 삽입
x = np.linspace(1,5,11)
y = x**2

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # 이동 좌표
ax2 = fig.add_axes([0.4, 0.5, 0.4, 0.3]) # 이동 좌표

ax1.plot(x , y)
ax2.plot(y , x)
ax1.set_title('plot1')
ax2.set_title('plot2')
fig
```

### 2. 객체 차트 (크기/해상도 조정)

```python
#길이크기조정, 해상도 조정

fig = plt.figure(figsize = (8,2), dpi = 60) # dpi로 해상도 조절 가능 높을수록 고해상도
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
fig
```

```python
#길이크기조정, 해상도 조정

fig = plt.figure(figsize = (8,5), dpi = 60) # dpi로 해상도 조절 가능 높을수록 고해상도
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```
