```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

# 커맨드뷰에서 차트 시연
%matplotlib inline
size = 50
x = np.linspace(1,10,10)
y = np.linspace(1,20,10)
y2= x**2
# 차트 꾸미기
plt.plot(x,y, "b--", label="first") #차트 선색 및 크기설정
plt.plot(x,y2, "r--", label="second") #차트 선색 및 크기설정
plt.legend(loc=1)
# 화살표 시작지점
plt.annotate('분리지점',xy=(2,2), xytext=(2,20),arrowprops=dict(color="blue", width=2))
```
